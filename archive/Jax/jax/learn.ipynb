```python
j3 = jnp.arange(10.)
```

```python
j3
```

```python
j4 = 2 * j3
```

```python
plt.plot(j3, j4)
plt.xlabel("j3")
plt.ylabel("j4")
plt.title("j3 vs j4")
```

```python
j5 = jnp.linspace(0, 10, 100)
j6 = jnp.sin(j5)
```

```python
plt.plot(j5, j6)
plt.xlabel("j5")
plt.ylabel("j6")
plt.title("j5 vs j6")
```

```python
def f(x : jnp.ndarray):
    return 2 * jnp.sin(x) * jnp.cos(x)

jit_f = jax.jit(f)
```

```python
%timeit f(j5)
```

```python
%timeit jit_f(j5)
```

```python
lst = [i for i in range(10)]
jax.vmap(jit_f)(jnp.array(lst))
```
